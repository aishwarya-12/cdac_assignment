```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tips.csv')
df.head(5)
```

```python
arr = np.array(df['total_bill'])
q1 = np.percentile(arr,25)
q3 = np.percentile(arr,75)
iqr = q3 - q1
lower_Fence = q1 - 1.5 *iqr
upper_Fence = q3 + 1.5 * iqr
```

```python
outliers = arr[(arr < lower_Fence) | (arr > upper_Fence) ]

print("Q1 ",q1)
print("Q2 ",np.median(arr))
print("Q3 ",q3)
print("IQR ", iqr)
print("Lower fence ", lower_Fence)
print("Upper fence ", upper_Fence)
print("Outliers ",outliers)
```

```python
#Boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data = df['total_bill'], orient = 'h')
plt.title("Box Plot of Total bill")
plt.xlabel("Amount")
plt.show()
```

```python
df1 = pd.read_csv('titanic.csv')
df1.head(5)
```

```python
titanic_arr = np.array(df1['age'])
q1 = np.percentile(titanic_arr,25)
q3 = np.percentile(titanic_arr,75)
iqr = q3 - q1
lower_Fence = q1 - 1.5 *iqr
upper_Fence = q3 + 1.5 * iqr

outliers = titanic_arr[(titanic_arr < lower_Fence) | (titanic_arr > upper_Fence) ]

print("Q1 ",q1)
print("Q2 ",np.median(titanic_arr))
print("Q3 ",q3)
print("IQR ", iqr)
print("Lower fence ", lower_Fence)
print("Upper fence ", upper_Fence)
print("Outliers ",outliers)
```

```python
five_point_summary = df1['age'].describe()[['min','25%','50%','75%','max']]
print(five_point_summary)

Q1 = df1['age'].quantile(0.25)
Q3 = df1['age'].quantile(0.75)
IQR = Q3-Q1
print("IQR",IQR)

lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
print("Lower_bound",lower_bound)
print("Upper_bound",upper_bound)

outliers = df1["age"][(df1['age']< lower_bound) | (df1['age'] > upper_bound) ]
print("Outliers ",outliers.values)
```

```python
#Boxplot
plt.figure(figsize=(10,6))
sns.boxplot(data = df1['age'], orient = 'h')
plt.title("Box Plot of Age")
plt.xlabel("Age")
plt.show()
```
