```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('airline_passenger_satisfaction.csv')
df.head(5)
```

```python
on_board_service = df['On-board Service']

mean_onboard = on_board_service.mean()
median_onboard = on_board_service.median()
skewness_onboard = on_board_service.skew()
kurtosis_onboard = on_board_service.kurt()

print(mean_onboard, median_onboard, skewness_onboard, kurtosis_onboard)
```

```python
plt.figure(figsize = (8,6))
sns.histplot(on_board_service, kde = True, color = "skyblue", bins = 5)
plt.axvline(mean_onboard, color = 'red', linestyle = '--', label = f'Mean : {mean_onboard:.2f}')
plt.axvline(median_onboard, color = 'black', linestyle = '--', label = f'Median : {median_onboard:.2f}')
plt.title(f'Distribution of On-Board Service\nSkewness: {skewness_onboard:.2f},Kurtosis:{kurtosis_onboard:.2f}')
plt.xlabel('On Board Service')
plt.ylabel('Frequency')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(on_board_service)
plt.tight_layout()
plt.show()
```

```python
#Cov
g1_name = 'First-time'
g2_name = 'Returning'

group1 = df[df['Customer Type'] == g1_name]['On-board Service']
group2 = df[df['Customer Type'] == g2_name]['On-board Service']

cov_group1 = group1.std() / group1.mean() * 100
cov_group2 = group2.std() / group2.mean() * 100

print(f"Coefficient of Variation (CoV) for {g1_name} : {cov_group1:.2f}%")
print(f"Coefficient of Variation (CoV) for {g2_name} : {cov_group2:.2f}%")
```
