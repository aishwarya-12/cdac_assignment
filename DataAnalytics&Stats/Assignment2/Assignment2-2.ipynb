```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('titanic.csv')
df.head(2)
```

```python
not_survived = len(df[df['survived'] == 0])
not_survived
```

```python
#Probability that a passenger dies (Marginal probability)
prob_not_survived = not_survived / (len(df['survived']))
print(f"Probability that passenger dies = {prob_not_survived:.2f}")
```

```python
no_of_males = len(df[df['gender'] == 'male'])
prob_male = no_of_males / (len(df['gender']))
print("Probability of male:",prob_male)

prob_survived = 1 - prob_not_survived
print("Probability of survival:",prob_survived)
```

```python
#Probability that a passenger was male and survived (Joint probability of independent events)
prob_male_survived = prob_male * prob_survived
print(f"Probability that passenger was male and survived = {prob_male_survived:.2f}")
```

```python
prob_of_female = 1 - prob_male
prob_of_female
```

```python
#Probability that a passenger survived given that she was a female (Conditional/Joint probability of dependent events)
no_survived_female = len(df[(df['survived'] == 1 ) & (df['gender'] == 'female')]) / len(df['survived'])
prob_survived_female = no_survived_female / prob_of_female
print(f"Probability that passenger survived was female = {prob_survived_female:.2f}")
```
