8. A courier company delivers parcels in a particular area in 30 minutes. A new GPS system installed has reduced the average time to 28 minutes, when 100 trucks were studied, with a sample standard deviation of 4.2 minutes. Run a two-tailed test and a one-tailed (left-tailed) test.


```python
mu0 =30
xbar = 28
n= 100
s = 4.2
alpha = 0.05
```

```python
z_crit_two_tailed = norm.ppf(1 - alpha/2) 
z_crit_one_tailed = norm.ppf(1 - alpha) 

print(f"Two tailed (alpha = 0.05) : +-{z_crit_two_tailed:.3f}")
print(f"One tailed (alpha = 0.05) : +{z_crit_one_tailed:.3f}")
```

```python
zscore = (xbar - mu0)/ (s / math.sqrt(n))
print(f"Z-score: {zscore:.3f}")
```

```python
p_two_tailed = 2*( 1 - norm.cdf(abs(zscore)))
p_one_tailed = norm.cdf(zscore)

if abs(zscore) > z_crit_two_tailed:
    two_tail_decision = "Reject H0"
else:
    two_tail_decision = "Fail to Reject H0"

if zscore < z_crit_one_tailed:
    one_tail_decision = "Reject H0"
else:
    one_tail_decision = "Fail to reject H0"

print("Two tailed test")
print(f" p-value = {p_two_tailed:.4f}")
print(f" Decision: {two_tail_decision}")

print("One tailed test")
print(f" p-value = {p_one_tailed:.4f}")
print(f" Decision: {one_tail_decision}")
```

9. A battery manufacturer claims that a battery which used run for 30 hours now runs for more than 30 hours due to new technology. 200 batteries were studied. Average battery time was found to be 29.5 hours with a standard deviation of 8 hours. Run a one-tailed (right-tailed) test.


```python
mu1 = 30 
n1 = 200
x_bar_b = 29.5
s1 = 8
alpha = 0.05
z_crit_one_tailed = norm.ppf(1 - alpha)
z_score_batteries = (x_bar_b - mu1)/ (s1 / math.sqrt(n1))
p_one_tailed_batteries = 1 - norm.cdf(z_score_batteries)
#one tailed Decision
if z_score_batteries > z_crit_one_tailed:
    one_tail_decision = "Reject H0 (Significant improvement)"
else:
    one_tail_decision = "Fail to reject H0 (No significant improvement)"
print(f"Z-score: {z_score_batteries:.3f}")
print("One tailed test;")
print(f" p-value = {p_one_tailed_batteries:.4f}")
print(f" Decision: {one_tail_decision}")
```

10. For the three combined Marathon datasets, extract the top 25 records with lowest finish time (Official Time column). Check the null hypothesis H0 = The average finishing time of the top 25 athletes is 130 minutes. Test it for α = 1%, 5%, and 10%.


```python
#T-test
df1 = pd.read_csv('marathon_results_2015.csv')
df2 =pd.read_csv('marathon_results_2016.csv')
df3 = pd.read_csv('marathon_results_2017.csv')
df = pd.concat([df1,df2,df3])
df.head()
```

```python
official_time = pd.to_timedelta(df["Official Time"])
official_time = official_time.dt.total_seconds() / 60
official_time = official_time.nsmallest(25)
avg_official_time = 130    #H0

t_statistic, p_value = stats.ttest_1samp(official_time,avg_official_time)
print(f"P-value: {p_value:.2f}")
```

```python
alpha = 0.05
if p_value > alpha:
    print("Do not reject H0")
else:
    print("Reject H0")
```

```python
alpha = 0.01
if p_value > alpha:
    print("Do not reject H0")
else:
    print("Reject H0")
```

```python
alpha = 0.1
if p_value > alpha:
    print("Do not reject H0")
else:
    print("Reject H0")
```

11. For combined Marathon datasets, find if younger (age <= 30) and older (age > 30) athletes have a similar average pace.

```python
df['Age'] = df['Age'].dropna()
df['Pace'] = df['Pace'].dropna()
df[['Hours','Minutes','Seconds']] =df['Pace'].str.split(':',expand =True)

df['Hours'] =df['Hours'].astype(int)
df['Minutes'] =df['Minutes'].astype(int)
df['Seconds'] =df['Seconds'].astype(int)
```

```python
df['pace_in_seconds'] = df['Hours'] * 3600 + df['Minutes']*60 + df['Seconds']
```

```python
younger = df[df['Age'] <= 30]['pace_in_seconds'].sample(n=50,replace=False)
older = df[df['Age'] > 30]['pace_in_seconds'].sample(n=50,replace=False)
```

```python
t_statistic,p_value = stats.ttest_ind(younger,older)

alpha=0.05
print("P-value:",p_value)
if p_value < alpha:
    print("Do not reject H0")
else:
    print("Reject H0")
```

12. For StudentsPerformance.csv, using ANOVA find if there is a significant difference in Maths marks depending on the level of parental education.

```python
#For StudentsPerformance.csv , using ANOVA
from scipy.stats import f_oneway
data_stud = pd.read_csv('StudentsPerformance.csv')
print(data_stud.columns)
```

```python
print(data_stud['parental level of education'].value_counts())
```

```python
s1 = data_stud['math score'][data_stud["parental level of education"] == "some college"]
s2 = data_stud['math score'][data_stud["parental level of education"] == "associate's degree"]
s3 = data_stud['math score'][data_stud["parental level of education"] == "high school"]
s4 = data_stud['math score'][data_stud["parental level of education"] == "some high school"]
s5 = data_stud['math score'][data_stud["parental level of education"] == "bachelor's degree"]
s6 = data_stud['math score'][data_stud["parental level of education"] == "master's degree"]

fsat, pval2 = f_oneway(s1, s2, s3, s4, s5, s6)

print(fsat)
print("P-value:",pval2)

alpha = 0.05

if pval2 < alpha :
    print("Reject H0")
else : 
    print("Do not reject H0")
```

13. Try to find if gender and smoking are related in the tips dataset.

```python
from scipy.stats import chi2_contingency

df_tips = pd.read_csv('tips.csv')
df_tips.columns

#create a contigency table
c_table1 = pd.crosstab(df_tips["gender"], df_tips["smoker"])
print(c_table1)

#Perform the chi-square test
chi2_stat1, p_val1, dof1, expected1 = chi2_contingency(c_table1)
print("chi-square statistic: ", chi2_stat1)
print("p-value : ", p_val1)
print("Degree of Freedom : ", dof1)

if p_val1 < 0.05:
    print("Reject H0")
else:
    print("Do not reject H0")
```

14. Consider diabetes dataset and find the relation of age with diabetes outcome.
(Hint: Group age into 3 categories: Young ≤ 30, Middle 31–50, Old > 50)


```python
diabetes = pd.read_csv('diabetes.csv')
diabetes.head()
```

```python
diabetes['AgeGroup'] = pd.cut(
    diabetes['Age'],
    bins=[0, 30, 50, diabetes['Age'].max()],
    labels=['young', 'middle', 'old'],
    right=True
)
#create a contigency table
c_table1 = pd.crosstab(diabetes['Outcome'], diabetes['AgeGroup'])
print(c_table1)

#Perform the chi-square test
chi2_stat1, p_val1, dof1, expected1 = chi2_contingency(c_table1)
print("chi-square statistic: ", chi2_stat1)
print("p-value : ", p_val1)
print("Degree of Freedom : ", dof1)

if p_val1 < 0.05:
    print("Reject H0")
else:
    print("Do not reject H0")
```
