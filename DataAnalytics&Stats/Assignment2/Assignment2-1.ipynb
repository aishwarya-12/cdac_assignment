Perform simple, systematic, stratified, cluster sampling.

Additional requirements:
(a) Ensure that we only have 10% of the records in each case. So, remove extra records, if necessary.
(b) For stratified, use strata as race.
(c) For cluster, use clusters as marital.status.

Then for overall data and for the four samples, find out the average education.num and average fnlwgt. Which of the four sampling methods is the closest to the overall average in the two cases?


```python
import pandas as pd
import numpy as np
```

```python
income_df = pd.read_csv('income.csv')
income_df.head()
```

```python
orignal_row_count = len(income_df)
print(orignal_row_count)
```

```python
#simple random sampling
simple_df = income_df.sample(frac = 0.1, random_state = 101)
print(simple_df.head())
simple_row_count = len(simple_df)
print("Simple row count", simple_row_count)
```

```python
#systematic random sampling
system_df = income_df.iloc[::10, :]
print(system_df.head())
print("Systematic Row Count", len(system_df))
```

```python
#stratified sampling on race
race = income_df['race'].unique()
print(race)
```

```python
total_white = len(income_df[income_df['race'] == 'White'])
total_black = len(income_df[income_df['race'] == 'Black'])
total_Asian = len(income_df[income_df['race'] == 'Asian-Pac-Islander'])
total_Amer = len(income_df[income_df['race'] == 'Amer-Indian-Eskimo'])
total_others = len(income_df[income_df['race'] == 'Other'])

sample_white = int(0.1 * total_white)
sample_black = int(0.1 * total_black)
sample_asian = int(0.1 * total_Asian)
sample_amer = int(0.1 * total_Amer)
sample_other = int(0.1 * total_others)

print( "White : ", total_white, sample_white)
print("Black : ", total_black, sample_black)
print("Asian : ", total_Asian, sample_asian)
print("American : ", total_Amer, sample_amer)
print("Others : ", total_others, sample_other)
```

```python
white_sample = income_df[income_df['race'] == 'White'].sample(n=sample_white, random_state = 101)
black_sample = income_df[income_df['race'] == 'Black'].sample(n=sample_black, random_state = 101)
asian_sample = income_df[income_df['race'] == 'Asian-Pac-Islander'].sample(n=sample_asian, random_state = 101)
amer_sample = income_df[income_df['race'] == 'Amer-Indian-Eskimo'].sample(n=sample_amer, random_state = 101)
other_sample = income_df[income_df['race'] == 'Other'].sample(n=sample_other, random_state = 101)
```

```python
stratified_sample = pd.concat([white_sample, black_sample, asian_sample, amer_sample, other_sample])
stratified_len = len(stratified_sample)
print(stratified_len)
```

```python
#Cluster sampling on marital status
marital_status = income_df['marital.status'].unique()
marital_status
```

```python
np.random.seed(42)
selected_marital_status = np.random.choice(marital_status, size = 3, replace = False)
print(selected_marital_status)
```

```python
sample1 = income_df[income_df['marital.status'] == 'Married-civ-spouse'].sample(frac = 0.17, random_state = 101)
sample2 = income_df[income_df['marital.status'] == 'Divorced'].sample(frac = 0.17, random_state = 101)
sample3 = income_df[income_df['marital.status'] == 'Married-spouse-absent'].sample(frac = 0.17, random_state = 101)

#m = cluster_df[cluster_df['marital.status']== 'Married-civ-spouse']
#d = cluster_df[cluster_df['marital.status']== 'Divorced']
#m1 = cluster_df[cluster_df['marital.status']== 'Married-spouse-absent']
#print(len(m),len(d),len(m1))
cluster_df = pd.concat([sample1,sample2,sample3])
cluster_df.head(10)
len(cluster_df)
```

```python
#Overall average
avg_overall_df_edu = round(income_df['education.num'].mean(),2)
avg_overall_df_fnlwgt = round(income_df['fnlwgt'].mean(),2)
print(f"Avg of education.num = {avg_overall_df_edu:.2f}")
print(f"Avg of fnlwgt = {avg_overall_df_fnlwgt:.2f}")
```

```python
#Simple sampling average
avg_simple_df_edu = round(simple_df['education.num'].mean(),2)
avg_simple_df_fnlwgt = round(simple_df['fnlwgt'].mean(),2)
print(f"Avg of education.num = {avg_simple_df_edu:.2f}")
print(f"Avg of fnlwgt = {avg_simple_df_fnlwgt:.2f}")
```

```python
#Systematic sampling average
avg_system_df_edu = round(system_df['education.num'].mean(),2)
avg_system_df_fnlwgt = round(system_df['fnlwgt'].mean(),2)
print(f"Avg of education.num = {avg_system_df_edu :.2f}")
print(f"Avg of fnlwgt = {avg_system_df_fnlwgt:.2f}")
```

```python
#Stratified sampling average
avg_strat_df_edu = round(stratified_sample['education.num'].mean(),2)
avg_strat_df_fnlwgt = round(stratified_sample['fnlwgt'].mean(),2)
print(f"Avg of education.num = {avg_strat_df_edu :.2f}")
print(f"Avg of fnlwgt = {avg_strat_df_fnlwgt:.2f}")
```

```python
#Cluster sampling average
avg_cluster_df_edu = round(cluster_df['education.num'].mean(),2)
avg_cluster_df_fnlwgt = round(cluster_df['fnlwgt'].mean(),2)
print(f"Avg of education.num = {avg_cluster_df_edu:.2f}")
print(f"Avg of fnlwgt = {avg_cluster_df_fnlwgt:.2f}")
```
