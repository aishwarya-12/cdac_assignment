```python
import pandas as pd
import numpy as np
from scipy.stats import zscore
```

```python
df = pd.read_csv('weight-height.csv')
df.columns
```

```python
weight_mean = df['Weight'].mean()
height_mean = df['Height'].mean()

df['weight_zscore'] = zscore(df['Weight'])
df['height_zscore'] = zscore(df['Height'])
```

```python
threshold = 3
weight_outliers = df[df['weight_zscore'].abs() > threshold ]
height_outliers = df[df['height_zscore'].abs() > threshold ]
all_outliers = pd.concat([weight_outliers, height_outliers]).drop_duplicates()
print(all_outliers)
```

```python
print("Number of Weight outliers : ", len(weight_outliers))
print("Number of height outliers  : ", len(height_outliers))
print("Number of Unique outliers : ", len(all_outliers))
```
