```python
#Pandas practice
import pandas as pd
```

```python
df = pd.read_csv('titanic.csv')
df.head(5)
```

```python
#Do EDA on dataset
print(df.tail(5))
print("\n",df.columns)
print("\n",df.describe())
```

```python
#Passengers between age 2 to 20
age_filter = df.query("age >= 2 & age <= 20")
print(age_filter)
```

```python
#Passengers where gender is female and class is 1
filter1 = df.query("gender == 'female' & pclass == 1")
filter1
```

```python
#Average age of female and male passengers
df['age'] = df['age'].fillna(0).astype('int64')
average_age = df.groupby('gender')['age'].mean()
average_age
```

```python
#Count number of passengers in each class
no_of_passengers = df.groupby('pclass')['pclass'].count()
no_of_passengers
```

```python
#Separate dataframes for survived and not survived passengers
df_survived = df.query('survived == 0')
print("Survived\n",df_survived)

df_notsurvived = df.query('survived == 1')
print("Not Survived\n",df_notsurvived)
```

```python
#No. of males, females in above dataframes
count_of_females_males = df_survived.groupby('gender')['gender'].count()
count_of_females_males
```

```python
count_of_females_males = df_notsurvived.groupby('gender')['gender'].count()
count_of_females_males
```

```python
#Sort passengers by age in descending order and display top 3
#df['age'] = df['age'].fillna(0).astype('int64')
sorted_data = df.sort_values(by = 'age', ascending = False)
sorted_data.head(3)
```
